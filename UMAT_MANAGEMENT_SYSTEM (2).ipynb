```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        print(f"Name: {self.name}, Age: {self.age}")


class Student(Person):
    def __init__(self, name, age, student_id):
        super().__init__(name, age)
        self.student_id = student_id

    def enroll_course(self, course_name):
        print(f"{self.name} (Student ID: {self.student_id}) has enrolled in {course_name}.")


class Lecturer(Person):
    def __init__(self, name, age, employee_id):
        super().__init__(name, age)
        self.employee_id = employee_id

    def teach_course(self, course_name):
        print(f"{self.name} (Employee ID: {self.employee_id}) is teaching {course_name}.")

    def display_info(self):
        super().display_info()
        print(f"Employee ID: {self.employee_id}")


print("--- Student Demo ---")
student1 = Student("OSEI BADU OLIVER", 20, "FOE.41.006.135.25")
student1.display_info()
student1.enroll_course("OBJECT ORIENTED PROGRAMMING")  # unique method

print("--- Lecturer Demo ---")
lecturer1 = Lecturer("MR MATTHEW COBBINAH", 45, "UMaT/E/1042")
lecturer1.display_info()
lecturer1.teach_course("OBJECT ORIENTED PROGRAMMING")
```
