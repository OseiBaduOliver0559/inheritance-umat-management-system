```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        print(f" Name: {self.name}, Age: {self.age} years old")

class Student(Person):
    def __init__(self, name, age, student_id, enroll_course):
        super().__init__(name, age)
        self.student_id = student_id
        self.enroll_course = enroll_course

    def display_info(self):
        super().display_info()
        print(f"A student has a Student ID:{self.student_id} and enroll in  {self.enroll_course} as course")

class Lecturer(Person):
    def __init__(self, name, age, employee_id, teacher_course):
        super().__init__(name, age)
        self.employee_id = employee_id
        self.teacher_course = teacher_course

    def display_info(self):
        super().display_info()
        print(f"A Lecturer has an Employee ID:{self.employee_id} and teaches {self.teacher_course}")

s1 = Student("OSEI BADU OLIVER", 21, "FOE.41.006.135.25", "Electrical and Electronic Eng.")
t1 = Lecturer("Mr Matthew Cobbinah", 35, "383898", "Object Oriented Programming")
s1.display_info()
t1.display_info()


```
